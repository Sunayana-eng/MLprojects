## Separating dependent(Target) and independent(Features)variable

```
x = distorted_dataset.iloc[:,:-1].values
y = distorted_dataset.iloc[:,-1].values
```

```
type(x)
```

```
y
```

```
y = y.reshape(len(y),1)
```

```
y
```

## Splitting dataset into training set and test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

## Feature Scaling 

```
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
y_train = sc_y.fit_transform(y_train)
x_test = sc_x.transform(x_test)
y_test = sc_y.transform(y_test)
```

# Training the Models and predicting Test sets

```
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
def KFold(model,X,Y,pred):
  cv_score = cross_val_score(model, X, Y, scoring='neg_mean_squared_error',cv = 5)
  cv_score_mean = np.abs(np.mean(cv_score))
  print("----Model Report----")
  print("MSE:",mean_squared_error(y_test, pred))
  print('CV Score:', cv_score)
  print('CV Score Mean:', cv_score_mean)
  print('R Squared:',r2_score(y_test,pred))
```

## Multiple Linear Regression

```
from sklearn.linear_model import LinearRegression
Lregressor = LinearRegression()
Lregressor.fit(x_train, y_train)
```

```
y_pred =sc_y.inverse_transform(Lregressor.predict(x_test))
np.set_printoptions(precision=2)
y_test = sc_y.inverse_transform(y_test)
```

```
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```
KFold(Lregressor,x_train,y_train,y_pred)
```

# Polynomial Regression

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x_train)
Pregressor = LinearRegression()
Pregressor.fit(x_poly,y_train)
```

```
y_pred_p = sc_y.inverse_transform(Pregressor.predict(poly_reg.transform(x_test)))
```

```

print(np.concatenate((y_pred_p.reshape(len(y_pred_p),1), y_test.reshape(len(y_test),1)),1))
```

```
KFold(Pregressor,x_poly,y_train,y_pred_p)
```

## SVR Regression Model

```
from sklearn.svm import SVR
Sregressor = SVR(kernel='rbf')
Sregressor.fit(x_train,y_train)
```

```
y_pred_svr =sc_y.inverse_transform(Sregressor.predict(x_test))
print(np.concatenate((y_pred_svr.reshape(len(y_pred_svr),1), y_test.reshape(len(y_test),1)),1))
```

```
KFold(Sregressor,x_train,y_train,y_pred_svr)
```

## Decision Tree Regressor

```
from sklearn.tree import DecisionTreeRegressor
Dregressor = DecisionTreeRegressor()
Dregressor.fit(x_train,y_train)
```

```
y_pred_dt = sc_y.inverse_transform(Dregressor.predict(x_test))
print(np.concatenate((y_pred_dt.reshape(len(y_pred_dt),1), y_test.reshape(len(y_test),1)),1))
```

```
KFold(Dregressor,x_train,y_train,y_pred_dt)
```

## Random Forest Tree Regression

```
from sklearn.ensemble import RandomForestRegressor
Rregressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
Rregressor.fit(x_train,y_train)
```

```
y_pred_rt = sc_y.inverse_transform(Rregressor.predict(x_test))
print(np.concatenate((y_pred_rt.reshape(len(y_pred_rt),1), y_test.reshape(len(y_test),1)),1))
```

```
KFold(Rregressor,x_train,y_train,y_pred_rt)
```
